15.Розв'яжіть  систему лінійних рівнянь A*x=b, де A -матриця коефіцієнтів, b - вектор членів

```
a = np.array([[1, 3, -4], [0, 2, -3], [-2, 2, 5]])
b = np.array([5, 3, 4])
x = np.linalg.solve(a, b)
print(x)
inv_a = np.linalg.inv(a)
x1 = inv_a.dot(b)
print(x1)
```

20. Сума матриць

```
a = np.matrix([[1, 2], [3, 4]])
b = np.matrix([[5, 6], [7, 8]])

print(np.add(a, b))
print(np.subtract(a, b))
print(np.multiply(a, b))
print(np.divide(a, b))
```

22.Сортування рядків або стопвців матриці

```
b = np.matrix([[5, 3, 4], [2, 3, 5], [3, 2, 3]])
print(b)
print(np.sort(b, axis=0))
print(np.sort(b, axis=1))
print(np.sort(b, axis=None))
```
