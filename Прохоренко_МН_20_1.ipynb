1.Створіть одновимірний масив (вектор) з сумами кожного числа та його
попереднього, починаючи з 1, та виведіть його значення.

```
import numpy as np

numbers = np.arange(1, 21)
sums = np.concatenate(([1], numbers[:-1] + numbers[1:]))
print(sums)
```

5.Створіть два одновимірних масиви розміром 4, заповніть їх випадковими цілими
числами від -10 до 10 та виконайте на них поелементні операції додавання, віднімання
та множення.

```
import numpy as np

a = np.random.randint(-10, 11, 4)
b = np.random.randint(-10, 11, 4)
print("Масив a:", a)
print("Масив b:", b)
print("Додавання:", a + b)
print("Віднімання:", a - b)
print("Множення:", a * b)
```

6.Створіть дві матриці 3x3, заповніть їх довільними цілими числами від 1 до 20 та
знайдіть їх добуток.

```
import numpy as np

matrix_a = np.random.randint(1, 21, size=(3, 3))
matrix_b = np.random.randint(1, 21, size=(3, 3))
result = matrix_a @ matrix_b

print(matrix_a)
print(matrix_b)
print(result)
```

12.Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в
діапазоні від 1 до 5 та знайдіть добуток елементів діагоналі другої матриці

```
import numpy as np

matrix1 = np.random.randint(1, 6, size=(3, 3))
matrix2 = np.random.randint(1, 6, size=(3, 3))
diagonal_product = np.prod(np.diag(matrix2))

print(matrix1)
print(matrix2)
print(diagonal_product)
```

15.Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в
діапазоні від -2 до 2 та знайдіть вектор-стовпчик, що містить кількість ненульових
елементів в кожному рядку матриці.

```
import numpy as np

matrix = np.random.uniform(-2, 2, size=(3, 3))
nonzero_counts = np.count_nonzero(matrix, axis=1).reshape(-1, 1)

print(matrix)
print(nonzero_counts)
```
